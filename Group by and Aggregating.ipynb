Group By and Aggregating

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/sherzod19/Downloads/Flavors.csv')
df
```

```python
df.groupby('Base Flavor')['Flavor Rating'].mean()
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.min()
```

```python
group_by_frame.sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count'], 
                               'Texture Rating':['mean', 'max', 'count'] })
```

```python
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count'], 
                               'Texture Rating':['mean', 'max', 'count']})
```

```python
group_by_frame.last()
```

```python
df
```

```python
df.groupby(["Base Flavor", "Liked"]).size().unstack()
```

```python
df.groupby('Flavor')['Total Rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('Liked')[["Flavor Rating", "Texture Rating", "Total Rating"]].agg(['mean', 'std'])
```

```python
df.groupby('Base Flavor')['Flavor'].nunique()
```

```python
df.groupby('Base Flavor')['Flavor Rating'].agg(lambda x: x.max() - x.min())
```

```python
df.groupby('Base Flavor')['Flavor Rating'].agg(lambda x: sorted(x, reverse=True)[1])
```

```python
df.groupby('Base Flavor')['Liked'].agg(lambda x: x.mode().iloc[0])
```
